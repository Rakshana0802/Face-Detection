```python
#import opencv
import cv2

#import matplotlib
import matplotlib.pyplot as plt
```

```python
def plottingImg(image):
    plt.axis('off')
    plt.figure(figsize=(15,15))
    plt.imshow(image)
    plt.show()
```

```python
def findFace(image):
    boxes = faceCascade.detectMultiScale(
        image,
        scaleFactor=1.2,
        minNeighbors=5,
        minSize=(50, 50)
        )
    return boxes

def drawBox(image, boxes):
    for (x,y,w,h) in boxes:
        cv2.rectangle(
            image, 
            (x,y), 
            (x+w,y+h), 
            (0,0,255), 
            5)
    return image
```
